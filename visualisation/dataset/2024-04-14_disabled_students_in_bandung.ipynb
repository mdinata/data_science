### Data Siswa Disabilitas Kota Bandung

https://opendata.bandung.go.id/dataset/jumlah-peserta-didik-disabilitas-di-sekolah-dasar-kota-bandung

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
file = 'dataset/jumlah_peserta_didik_disabilitas_di_sekolah_dasar_kota_bandung.xlsx'
data = pd.read_excel(file)
````

Get an overview of the data by using ```info()```

````python
data.info()
````

```info()``` shows the column name, count of non-null data in each column and data type. int64 Dtype refers to 64 bit integer data (real number). Object Dtype refers to text / string type.  

To get the preview of first 5 data use ```head()``` and last 5 data use ```tail()```

````python
data.head()
````

````python
data.tail()
````

## Single Column

````python
data.groupby('tahun')['jumlah_siswa_disabilitas'].sum()
````

````python
data.groupby('jenis_kelamin')['jumlah_siswa_disabilitas'].sum()
````

````python
data.groupby('status_sekolah')['jumlah_siswa_disabilitas'].sum()
````

## Multi-column

````python
data.groupby(['jenis_kelamin','tahun'])['jumlah_siswa_disabilitas'].sum()
````

````python
data.groupby(['status_sekolah','tahun'])['jumlah_siswa_disabilitas'].sum()
````

## Visualisasi dengan Grafis

### Single Column

````python
result = data.groupby('tahun')['jumlah_siswa_disabilitas'].sum()
````

````python
result.plot(kind='bar')
plt.title('Jumlah Siswa Disabilitas Kota Bandung Tahunan')
plt.show()
````

### Multi-Column

````python
result = data.groupby(['jenis_kelamin','tahun'])['jumlah_siswa_disabilitas'].sum()
result.plot(kind='bar')
plt.title('Jumlah Siswa Disabilitas Kota Bandung\n Laki-laki dan Perempuan\n Kota Bandung')
plt.show()
````

````python
result = data.groupby(['status_sekolah','tahun'])['jumlah_siswa_disabilitas'].sum()
result.plot(kind='bar')
plt.title('Jumlah Siswa Disabilitas Kota Bandung\n Sekolah Swasta dan Negeri\n Kota Bandung')
plt.show()
````

````python
result = data.groupby(['jenis_kelamin'])['jumlah_siswa_disabilitas'].sum()
result.plot(kind='pie')
````

## Display Multiple Charts with subplots

````python
fig, ax = plt.subplots(1, 4, figsize=(18, 4))

result1 = data.groupby(['status_sekolah'])['jumlah_siswa_disabilitas'].sum()
result1.plot(ax = ax[3],title='Proporsi Siswa Disabilitas\nKota Bandung\n2021-2023',kind='pie')

result2 = data.groupby(['status_sekolah','tahun'])['jumlah_siswa_disabilitas'].sum()
result2.plot(ax = ax[0],title='Total Siswa Disabilitas\nBerdasarkan Tipe Sekolah\nKota Bandung\n2021-2023',kind='bar')

result3 = data.groupby(['tahun','jenis_kelamin'])['jumlah_siswa_disabilitas'].sum()
result3.plot(ax = ax[1],title='Total Siswa\nBerdasarkan Jenis Kelamin\nPer Tahun\nKota Bandung',kind='bar')

result4 = data.groupby(['jenis_kelamin'])['jumlah_siswa_disabilitas'].sum()
result4.plot(ax = ax[2],title='Total Siswa Disabilitas\nBerdasarkan Jenis Kelamin\nKota Bandung\n2021-2023',kind='bar')

plt.show()
````
